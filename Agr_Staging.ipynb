```python
import pandas as pd
from sqlalchemy import create_engine

# Connect to staging_db
server = "localhost"
database = "agr_staging_db"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
engine = create_engine(connection_string)

# Load Excel file
df1 = pd.read_csv(r"F:\رواد مصر تراك تحليل البيانات\Tasks & Practices\Agriculture project\Irrigated Land_data_en_2-24-2026.csv", encoding='latin-1')
df2=pd.read_csv(r"F:\رواد مصر تراك تحليل البيانات\Tasks & Practices\Agriculture project\FAOSTAT_data_en_2-23-2026.csv", encoding='latin-1')
df=pd.concat([df2,df1])

# Insert into staging table
df.to_sql("irrigated_land_data", engine, if_exists="replace", index=False)
print("CSV data loaded into staging_db!")
```
